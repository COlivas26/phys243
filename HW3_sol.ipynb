```python
import gc
import random
import numpy as np
import matplotlib.pyplot as plt
import buffonNeedle as bn
```

# Numerical Simulation

```python
needleCounts = list(map(lambda x: int(x), np.linspace(100, 10**4, 3)))
```

## Part 2 - Short Needle

```python
bShortNeedleSim = bn.BuffonNeedleSimulation()
```

```python
# l and t values such that l <= t (sn = short needle)
snT = [10 for i in range(10)]
snL = [i+1 for i in range(10)]
bShortNeedleSim.simulate(snT, snL, needleCounts, True)
```

## Part 3 - Long Needle

```python
bLongNeedleSim = bn.BuffonNeedleSimulation()
```

```python
# l and t values such that l > t (ln = long needle)
lnT = [2 for i in range(20)]
lnL = [i+1 for i in range(20)]
bLongNeedleSim.simulate(lnT, lnL, needleCounts, True)
```

## Part 4 & 5 - Plot using Numerical Simulation & Analytic Formula

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))
plt.close() # Use close() to prevent fig to display empty plots in Jupyter Notebook
```

### Plot Short Needle

```python
# Clear the axis so that we won't be adding the same plots again
axes[0].clear()
axes[1].clear()

# Plot numerical simulations for short needle
bShortNeedleSim.plot(isNumericalSim=True, axis=axes[0])
bShortNeedleSim.plot(isNumericalSim=False, axis=axes[1])
```

### Plot Long Needle

```python
# Clear the axis so that we won't be adding the same plots again
axes[0].clear()
axes[1].clear()

# Plot analytic formulas for long needle
bLongNeedleSim.plot(isNumericalSim=True, axis=axes[0])
bLongNeedleSim.plot(isNumericalSim=False, axis=axes[1])
```

For short needle, as the (l/t) ratio increases, the probability reaches to $\frac{2l} {t\pi}$ and the graph becomes linear.

For long needle, once the (l/t) ratio is greater than 1 the probability keeps getting closer to 1 as (l/t) increases.

## Extra - Combine Short & Long Needle

```python
combinedBnSim = bn.BuffonNeedleSimulation()
```

```python
combinedT = snT + lnT
combinedL = snL + lnL
combinedBnSim.simulate(combinedT, combinedL, needleCounts)
```

```python
# Clear the axis so that we won't be adding the same plots again
axes[0].clear()
axes[1].clear()

# Plot analytic formulas for short and long needle
combinedBnSim.plot(isNumericalSim=True, axis=axes[0])
combinedBnSim.plot(isNumericalSim=False, axis=axes[1])
```
